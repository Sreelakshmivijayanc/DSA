```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
data=pd.read_csv(r'C:\Users\SREELAKSHMI\Downloads\StudentsPerformance.csv')
data
```

```python
data['average score']=(data['math score']+data['reading score']+data['writing score'])/3 #new colomn added
data
```

3. Who scores the most on average for math, reading and writing based on
● Gender
● Test preparation course



```python
#Female students have a higher average score than male studens. Students who have completed the test preparation course have higher marks.
```

```python
df=data.groupby(["gender","test preparation course"])[["average score"]].mean()
df
```

```python
sns.barplot(x='test preparation course',y='average score',data=data,hue='gender')
plt.title('AVERAGE SCORE')
plt.show()
```

```python
df2=data.groupby(["gender","test preparation course"])[["reading score","math score","writing score"]].mean()
```

```python
df2
```

```python
sns.barplot(x='test preparation course',y='math score',data=data,hue='gender')
plt.title('MATH SCORE')
plt.show()
```

```python
sns.barplot(x='test preparation course',y='reading score',data=data,hue='gender')
plt.title('READING SCORE')
plt.show()
```

```python
sns.barplot(x='test preparation course',y='writing score',data=data,hue='gender')
plt.title('WRITING SCORE')
plt.show()
```
